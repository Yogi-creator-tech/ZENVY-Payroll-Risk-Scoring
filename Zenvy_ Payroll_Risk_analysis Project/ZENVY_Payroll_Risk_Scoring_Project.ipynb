```python
# Converting Workload column to numeric 

df["Work load Average/day "] = pd.to_numeric(
    df["Work load Average/day "].astype(str).str.replace(",", ""),
    errors="coerce"
)

# Filling missing values

num_cols = df.select_dtypes(include=["float64","int64"]).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Feature Engineering 
df["absence_flag"] = (df["Absenteeism time in hours"] > 0).astype(int)

df["high_absence_flag"] = (
    df["Absenteeism time in hours"] >
    df["Absenteeism time in hours"].median()
).astype(int)

df["absence_ratio"] = df["Absenteeism time in hours"] / (df["Service time"] + 1)
```

## Feature Engineering â€“ Attendance & Leave Behaviour Features

New features were engineered from absenteeism hours and service time
to capture behavioural risk patterns that are not directly visible in the raw dataset.

| Feature | Description | Risk Relevance |
|--------|----------|-------------|
| absence_flag | 1 if employee was absent at least once | identifies leave occurrence |
| high_absence_flag | 1 if absenteeism hours are above median | detects frequent or long absences |
| absence_ratio | absenteeism hours / service time | measures attendance stability |

Employees with consistently high absenteeism ratios are more likely
to generate payroll anomalies, manual salary corrections or irregular
payroll adjustment records.

Therefore, attendance-based features are important inputs for the
Payroll Risk Scoring System.


```python
np.random.seed(42)

# Making a new column which is based on service time (experience band)
df["previous_salary"] = (
    25000 + (df["Service time"] * 800) +
    np.random.randint(1000, 4000, len(df))
)

# Current salary coloumn after increment (simulated payroll change)

df["current_salary"] = df["previous_salary"] * np.random.uniform(1.0, 1.25, len(df))

# Salary change percentage
df["salary_change_pct"] = (
    (df["current_salary"] - df["previous_salary"]) /
    df["previous_salary"]
) * 100

df[["Service time","previous_salary","current_salary","salary_change_pct"]].head()
```

```python
df["risk_label"] = (
    (df["high_absence_flag"] == 1) &
    (df["salary_change_pct"] > 15) &
    (df["absence_ratio"] > df["absence_ratio"].median())
).astype(int)

df["risk_label"].value_counts()
```

```python
df["risk_label"].value_counts().plot(kind="bar")

plt.xticks([0,1], ["Low Risk","High Risk"], rotation=0)
plt.ylabel("Number of Records")
plt.title("Payroll Risk Class Distribution")
plt.show()
```

## Payroll Risk Label Creation & Class Distribution

A rule-based risk label was created using three behavioural indicators:

1) High absenteeism hours  
2) Abnormal salary change percentage  
3) Unstable attendance ratio  

Employees who satisfied all three conditions were classified as **High-Risk (1)**,  
while the remaining employees were classified as **Low-Risk (0)**.

The class distribution shows that most payroll records belong to the
Low-Risk category, while only a smaller subset falls under High-Risk cases.

This behaviour is realistic in real-world payroll systems, where
risk events such as unusual salary jumps, excessive absenteeism or
manual payroll adjustments occur only in a minority of employee records.

Since the dataset is slightly imbalanced, evaluation metrics such as  
**Recall and ROC-AUC are more meaningful than Accuracy**, because
missing a high-risk employee record is more costly than a few false positives.


```python
plt.figure(figsize=(8,6))

corr = df[[
    "absence_ratio",
    "high_absence_flag",
    "salary_change_pct",
    "risk_label"
]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix â€” Engineered Features vs Risk Label")
plt.show()
```

## Correlation Analysis â€” Engineered Features vs Risk Label

From the correlation heatmap, the following observations can be made:

â€¢ `high_absence_flag` has the highest positive correlation with `risk_label`
  which means employees with frequent or long absenteeism are more likely
  to fall in the High-Risk payroll category.

â€¢ `salary_change_pct` also shows a moderate positive correlation with `risk_label`
  indicating that unusual or higher salary change percentages contribute to
  payroll risk behaviour.

â€¢ `absence_ratio` has a weaker but positive correlation with `risk_label`
  meaning employees with unstable attendance patterns still add some risk impact.

Overall, the risk label is influenced by a combination of absenteeism behaviour
and salary change variation, which supports the engineered feature design.


```python
from sklearn.model_selection import train_test_split

features = [
    "absence_ratio",
    "high_absence_flag",
    "salary_change_pct"
]

X = df[features]
y = df["risk_label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1442
)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
models = {

    "SVM (RBF Kernel)": {
        "model": SVC(probability=True),
        "params": {
            "C": [0.5, 1, 5, 10],
            "gamma": ["scale", 0.1, 0.01],
            "kernel": ["rbf"]
        }
    },

    "Random Forest": {
            "model": RandomForestClassifier(random_state=42),
            "params": {
                "n_estimators": [200, 300, 400],
                "max_depth": [None, 6, 10, 14],
                "min_samples_split": [2, 4, 6],
                "min_samples_leaf": [1, 2]
            }
    },

    "Gradient Boosting": {
        "model": GradientBoostingClassifier(random_state=42),
        "params": {
            "learning_rate": [0.01, 0.05, 0.1],
            "n_estimators": [100, 150, 200],
            "subsample": [0.8, 1.0],
            "max_depth": [2, 3]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score

tuned_results = []

for name, mp in models.items():

    print("\n--------------------------------------------------")
    print(f" MODEL TRAINING : {name}")
    print("--------------------------------------------------")

    gs = GridSearchCV(
        estimator = mp["model"],
        param_grid = mp["params"],
        scoring = "roc_auc",
        cv = 5,
        n_jobs = -1
    )

    gs.fit(X_train, y_train)

    best_model = gs.best_estimator_
    y_pred = best_model.predict(X_test)
    y_prob = best_model.predict_proba(X_test)[:,1]

    auc = roc_auc_score(y_test, y_prob)
    tuned_results.append([name, auc, gs.best_params_])

    print(f"\n Best ROC-AUC Score : {auc:.4f}")
    print("\n Best Hyperparameters :")
    for k,v in gs.best_params_.items():
        print(f"   â€¢ {k} = {v}")

    print("\n Classification Report :\n")
    print(classification_report(y_test, y_pred))
```

```python
comparison_df = pd.DataFrame(
    tuned_results,
    columns=["Model", "ROC_AUC", "Best Parameters"]
)

comparison_df
```

```python
final_model = GradientBoostingClassifier(
    learning_rate=0.05,
    n_estimators=100,
    max_depth=2,
    subsample=0.8,
    random_state=42
)

final_model.fit(X_train, y_train)

importances = final_model.feature_importances_

df_final = pd.DataFrame({
    "Feature": features,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)
df_final
```

## Feature Importance Analysis

The Gradient Boosting model was used to compute feature importance values.
The results show that the risk score is primarily influenced by:

1) absence_ratio  â†’  0.41 importance
   Employees with higher absenteeism relative to service time have a much
   higher probability of being classified as High-Risk.

2) salary_change_pct  â†’  0.36 importance
   Sudden or unusually high salary change percentages contribute strongly
   to payroll risk behaviour.

3) high_absence_flag  â†’  0.21 importance
   Frequent or long-duration absenteeism still plays an important role,
   but its effect is lower compared to cumulative attendance ratio.

Overall, the model indicates that payroll risk is driven by a combination of:
â€¢ attendance instability
â€¢ behavioural leave patterns
â€¢ abnormal salary revision events

This validates the engineered features and aligns with the project objective
of identifying high-risk payroll records based on behavioural indicators.
